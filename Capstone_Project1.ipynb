## **Type Conversion**

```python
df['Age'] = df['Age'].astype('float64')
```

```python
df.info()
```

```python
def convertMtoY(months):
  return months/12
```

```python
df['Age'] = df['Age'].apply(convertMtoY)
df['Age']
```

## **Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
Normalize = MinMaxScaler()
df['Age'] = Normalize.fit_transform(df['Age'].values.reshape(-1,1))
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
standardization = StandardScaler()
df['Age'] = standardization.fit_transform(df['Age'].values.reshape(-1,1))
df.describe().round(2) #also df['Age'] for only Age column description round(n) rounds upto n decimals
```
