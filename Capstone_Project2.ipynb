```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv('stocks_daily_prices.csv')
df2 = pd.read_csv('stocks_daily_returns.csv')
display(df1)
display(df2)
```

# **Line Plot**

```python
df1.plot(x='Date',linewidth=0.4,figsize=(12,7))
plt.grid()
```

```python
df1.plot(x='Date',y=['AAPL','JPM','PG','UAL'],subplots=True,linewidth=0.5,figsize=(15,15),grid=True)
# plt.legend(loc='upper right')
```

```python
plt.figure(figsize=(20,4))
plt.subplot(1,4,1)
plt.plot(df1['AAPL'])
plt.title('AAPL')
plt.grid()
plt.subplot(1,4,2)
plt.plot(df1['JPM'])
plt.title('JPM')
plt.grid()
plt.subplot(1,4,3)
plt.plot(df1['PG'])
plt.title('PG')
plt.grid()
plt.subplot(1,4,4)
plt.plot(df1['UAL'])
plt.title('UAL')
plt.grid()
```

```python
df2.plot.scatter(x='AAPL',y='JPM')
```

```python
sns.scatterplot(x=df2['AAPL'],y=df2['JPM'])
```

```python
arr = pd.DataFrame(data={'pct':[70,10,10,10]},index=['AAPL','JPM','PG','UAL'])
arr
```

```python
arr.plot.pie(y='pct',explode=(0.2,0,0,0),figsize=(12,7),shadow=True)
plt.legend(loc='upper right')
```

```python
ualmu = df2['UAL'].mean().round(2)
ualsigma = df2['UAL'].std().round(2)
pgmu = df2['PG'].mean().round(2)
pgsigma = df2['PG'].std().round(2)
df2.plot.hist(bins=40,figsize=(12,7))
plt.title('ualmu:'+str(ualmu)+' ualsigma:'+str(ualsigma)+'  pgmu:'+str(pgmu)+'  pgsigma'+str(pgsigma))
```

```python
correlations = df2.corr()
sns.heatmap(correlations,annot=True)
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111,projection='3d')

x = df2['AAPL'].tolist()
y = df2['JPM'].tolist()
z = df2['UAL'].tolist()

ax.scatter(x,y,z,s=500,c='green',marker='P')
```
